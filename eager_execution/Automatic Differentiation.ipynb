This code snippet demonstrates how automatic differentiation is done under eager execution mode.

```python
# setup tensorflow with eager execution
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
tf.enable_eager_execution()

tfe = tf.contrib.eager
```

### 1. Computing Derivatives

Below is an example that demonstrates the API for automatic differentiation given a function pointer.

```python
from math import pi
def f(x):
    return tf.square(tf.sin(x))
```

```python
x = np.arange(0,10,0.1, dtype='float32')
y = f(x).numpy()

plt.plot(x,y)
plt.title('f=(sin(x))**2')
plt.show()
```

```python
# evaluate derivative or gradient of function f at each of the x-values
grad_f = tfe.gradients_function(f)(x)[0]

y_ = grad_f.numpy()
plt.plot(x,y_)
plt.title('df/dx=2sin(x)cos(x)')
plt.show()
```

#### 1.1 Computing Higher-Order Derivatives

The gradients_function method can be applied multiple times on a function to get higher derivatives.

```python
# evluate 2nd derivative of function f
grad_f  = tfe.gradients_function(f)
grad_f2 = tfe.gradients_function(grad_f)
grad_f3 = tfe.gradients_function(grad_f2)

y  = f(x).numpy()
d1 = grad_f(x)[0].numpy()
d2 = grad_f2(x)[0].numpy()
d3 = grad_f3(x)[0].numpy()

plt.plot(x,y)
plt.title('f(x)')
plt.show()
plt.plot(x,d1)
plt.title('df/dx')
plt.show()
plt.plot(x,d2)
plt.title('d2f/dx2')
plt.show()
plt.plot(x,d3)
plt.title('d3f/dx3')
plt.show()
```
